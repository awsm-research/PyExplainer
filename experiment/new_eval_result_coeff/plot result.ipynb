```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

# Plot explanation result

```python
openstack_rf = pd.read_csv('./RQ3_openstack_RF_explanation_eval_split_rulefit_condition.csv')
qt_rf = pd.read_csv('./RQ3_qt_RF_explanation_eval_split_rulefit_condition.csv')
result_rf = pd.concat([openstack_rf, qt_rf])
result_rf['global_model'] = 'RF'

# result_rf.to_csv('./RQ3_RF_explanation_eval_split_rulefit_condition.csv',index=False)
openstack_lr = pd.read_csv('./RQ3_openstack_LR_explanation_eval_split_rulefit_condition.csv')
qt_lr = pd.read_csv('./RQ3_qt_LR_explanation_eval_split_rulefit_condition.csv')
result_lr = pd.concat([openstack_lr, qt_lr])
result_lr['global_model'] = 'LR'

all_result = pd.concat([result_rf, result_lr])
```

```python
fig, axs = plt.subplots(2,2, figsize=(8,8))

axs[0,0].set_title('RF')
axs[0,1].set_title('LR')

# plt.ylim(0,100)
sns.boxplot(data=result_rf, x='project', y='precision', hue='method', ax=axs[0,0])
sns.boxplot(data=result_rf, x='project', y='recall', hue='method', ax=axs[1,0])
sns.boxplot(data=result_lr, x='project', y='precision', hue='method', ax=axs[0,1])
sns.boxplot(data=result_lr, x='project', y='recall', hue='method', ax=axs[1,1])

plt.show()

display(all_result[['global_model', 'project','method', 'precision','recall']]
        .groupby(['global_model','project','method']).describe())
```

# Plot result of RQ4_combined_prob_from_guidance_ppick.csv

```python
result_df = pd.read_csv('RQ4_combined_prob_from_guidance_ppick.csv')

result_df['prob_revise_percent_diff'] = result_df['probOg'] - result_df['probRevisedPercent']
result_df['prob_revise_sd_diff'] = result_df['probOg'] - result_df['probRevisedSD']

result_rf = result_df[result_df['model']=='RF']
result_lr = result_df[result_df['model']=='LR']

print(result_df.head())

fig, axs = plt.subplots(2,2, figsize=(12,8))

axs[0,0].set_title('RF')
axs[0,1].set_title('LR')

# plt.ylim(0,100)
sns.boxplot(data=result_rf, x='project', y='prob_revise_percent_diff', hue='gtype', ax=axs[0,0])
sns.boxplot(data=result_rf, x='project', y='prob_revise_sd_diff', hue='gtype', ax=axs[1,0])
sns.boxplot(data=result_lr, x='project', y='prob_revise_percent_diff', hue='gtype', ax=axs[0,1])
sns.boxplot(data=result_lr, x='project', y='prob_revise_sd_diff', hue='gtype', ax=axs[1,1])

plt.show()

display(result_df[['model', 'project','gtype', 'prob_revise_percent_diff','prob_revise_sd_diff']]
        .groupby(['model','project','gtype']).describe())
```

```python
result_df['pred_original'] = np.round(result_df['probOg']).astype(bool)
result_df['pred_revise_percent_reverse'] = np.round(result_df['probRevisedPercent']).astype(bool)
result_df['pred_revise_sd_reverse'] = np.round(result_df['probRevisedSD']).astype(bool)
result_df['is_pred_revise_percent_reverse'] = (result_df['pred_original']==True) &  (result_df['pred_revise_percent_reverse']==False) 
result_df['is_pred_revise_sd_reverse'] = (result_df['pred_original']==True) &  (result_df['pred_revise_sd_reverse']==False) 

summary = result_df.groupby(['project','gtype','model']).sum()
display(summary[['is_pred_revise_percent_reverse', 'is_pred_revise_sd_reverse']])
    
# print(result_df)
```
